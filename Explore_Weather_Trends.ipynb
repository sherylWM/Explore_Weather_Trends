```python
import pandas as pd
```

```python
globalTempDF = pd.read_csv('/Applications/SHERYL 2019/Udacity Data Analyst Nano Degree/globaldata.csv')
cityTempDF = pd.read_csv('/Applications/SHERYL 2019/Udacity Data Analyst Nano Degree/citytemp.csv')
```

```python
#checking first 5 values of my city "Indianapolis" dataframe
cityTempDF.head(5)
```

```python
#checking if any value is null for my city "Indianapolis" dataframe
#column avg_temp has 5 null values for the city Indianapolis
cityTempDF.isnull().sum()
```

```python
#checking if any value is null for global temperature dataframe
#No null values in the global temperature dataframe
globalTempDF.isnull().sum()
```

```python
#To handle missing values in avg_temp column of cityTempDF I decided to select the mean imputation technique
#meanTemp is the average temperature of all values in avg_temp column
meanTemp = cityTempDF['avg_temp'].mean()
```

```python
#filling all NaN values in avg_temp column with the meanTemp value calculated earlier
cityTempDF['avg_temp'].fillna(meanTemp, inplace = True)
```

```python
#calculating 5 year moving average for avg_temp column for my city "Indianapolis" dataframe
cityTempDF['5yr_moving_avg_temp'] = cityTempDF['avg_temp'].rolling(window=5).mean()

#calculating 7 year moving average for avg_temp column for my city "Indianapolis" dataframe
cityTempDF['7yr_moving_avg_temp'] = cityTempDF['avg_temp'].rolling(window=7).mean()

#calculating 10 year moving average for avg_temp column for my city "Indianapolis" dataframe
cityTempDF['10yr_moving_avg_temp'] = cityTempDF['avg_temp'].rolling(window=10).mean()
```

```python
#calculating 5 year moving average for avg_temp column for global dataframe
globalTempDF['5yr_moving_avg_temp'] = globalTempDF['avg_temp'].rolling(window=5).mean()

#calculating 7 year moving average for avg_temp column for global dataframe
globalTempDF['7yr_moving_avg_temp'] = globalTempDF['avg_temp'].rolling(window=7).mean()

#calculating 10 year moving average for avg_temp column for global dataframe
globalTempDF['10yr_moving_avg_temp'] = globalTempDF['avg_temp'].rolling(window=10).mean()
```
